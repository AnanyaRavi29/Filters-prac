```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import freqz

def moving_average_filter(signal, window_size):
    filtered_signal = np.zeros_like(signal)
    for i in range(len(signal)):
        if i >= window_size - 1:
            filtered_signal[i] = np.mean(signal[i-window_size+1:i+1])
    return filtered_signal

def compute_frequency_response(window_size, sampling_frequency):
    b = np.ones(window_size) / window_size
    w, h = freqz(b, worN=8000)
    return 0.5 * sampling_frequency * w / np.pi, np.abs(h)

ECG = pd.read_csv('ECG3-1.csv')
time = ECG.iloc[:, 0].values
ecg_signal = ECG.iloc[:, 1].values

sampling_frequency = 100
window_size = 5

# Apply the first moving average filter
first_filtered_signal = moving_average_filter(ecg_signal, window_size)

# Apply the second moving average filter
second_filtered_signal = moving_average_filter(first_filtered_signal, window_size)

# Plot original and filtered ECG signals separately
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(time, ecg_signal, label='Original ECG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original ECG Signal')
plt.legend()
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(time, first_filtered_signal, label='Filtered ECG Signal (1st Filter)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Filtered ECG Signal (1st Filter)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(time, second_filtered_signal, label='Filtered ECG Signal (2nd Filter)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Filtered ECG Signal (2nd Filter)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Plot original and filtered ECG signals together
plt.figure(figsize=(10, 6))
plt.plot(time, ecg_signal, label='Original ECG Signal')
plt.plot(time, second_filtered_signal, label='Filtered ECG Signal (2nd Filter)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original and Filtered ECG Signals')
plt.legend()
plt.grid()
plt.show()

# Compute and plot the frequency response of the second filter
frequency_bins, magnitude_response = compute_frequency_response(window_size, sampling_frequency)

plt.figure(figsize=(10, 6))
plt.plot(frequency_bins, magnitude_response, 'b')
plt.title("Frequency Response of Second Moving Average Filter")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import freqz

def compute_derivative_filter(window_size):
    return np.array([-1, -2, 0, 2, 1]) / (2 * window_size ** 2)

def apply_filter(signal, filter_coeffs):
    filtered_signal = np.zeros(len(signal))
    for i in range(len(signal)):
        if i >= len(filter_coeffs) - 1:
            filtered_signal[i] = np.dot(signal[i-len(filter_coeffs)+1:i+1], filter_coeffs)
    return filtered_signal

def compute_frequency_response(filter_coeffs, signal_length):
    w, h = freqz(filter_coeffs, worN=signal_length)
    return w, np.abs(h)

def plot_results(time, signals, freq_responses, titles):
    num_plots = len(signals)
    plt.figure(figsize=(12, 8))
    for i in range(num_plots):
        plt.subplot(2, num_plots, i+1)
        plt.plot(time[i], signals[i])
        plt.title(titles[i] + ' Filtered Signal')
        plt.xlabel('Time (s)')
        plt.ylabel('Amplitude')

        plt.subplot(2, num_plots, num_plots+i+1)
        plt.plot(0.5 * 1000 * freq_responses[i][0] / np.pi, np.abs(freq_responses[i][1]))
        plt.title('Frequency Response of Filter for ' + titles[i])
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Gain')

    plt.tight_layout()
    plt.show()

# Read ECG signal 1 from CSV file
ecg_1 = pd.read_csv('ECG1.csv')
time_1 = ecg_1.iloc[:, 0].values
signal_1 = ecg_1.iloc[:, 1].values

# Read ECG signal 2 from CSV file
ecg_2 = pd.read_csv('ECG2.csv')
time_2 = ecg_2.iloc[:, 0].values
signal_2 = ecg_2.iloc[:, 1].values

# Compute filter coefficients
window_size = 5
filter_coeffs_1 = compute_derivative_filter(window_size)
filter_coeffs_2 = compute_derivative_filter(window_size)

# Apply filter to signals
filtered_signal_1 = apply_filter(signal_1, filter_coeffs_1)
filtered_signal_2 = apply_filter(signal_2, filter_coeffs_2)

# Compute frequency responses
w_1, h_1 = compute_frequency_response(filter_coeffs_1, len(signal_1))
w_2, h_2 = compute_frequency_response(filter_coeffs_2, len(signal_2))

# Plot results
plot_results([time_1, time_2],
             [filtered_signal_1, filtered_signal_2],
             [(w_1, h_1), (w_2, h_2)], ['ECG 1', 'ECG 2'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz
import pandas as pd

def apply_filter(input_signal, filter_coeffs):
    filtered_signal = np.zeros(len(input_signal))
    for i in range(len(input_signal)):
        for j in range(len(filter_coeffs)):
            if i - j >= 0:
                filtered_signal[i] += filter_coeffs[j] * input_signal[i - j]
    return filtered_signal

# Read ECG signal from CSV file
ecg_data = pd.read_csv('ECG3-1.csv')

# Extract time and ECG signal values
time = ecg_data.iloc[:, 0].values
ecg_signal = ecg_data.iloc[:, 1].values

# Define sampling frequency
sampling_frequency = 500  # Assuming sampling frequency is 500Hz

# Define coefficients for third-order derivative filter
filter_coeffs = [1, -3, 3, -1]

# Apply third-order derivative filter
filtered_ecg_signal = apply_filter(ecg_signal, filter_coeffs)

# Compute frequency response of the filter
w, h = freqz(filter_coeffs, worN=8000)

# Plot original and filtered ECG signals
plt.figure(figsize=(10, 6))
plt.plot(time, ecg_signal, label='Original ECG Signal')
plt.plot(time, filtered_ecg_signal, label='Filtered ECG Signal (Third-order Derivative)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('ECG Signal Before and After Filtering')
plt.legend()
plt.grid()

# Plot frequency response of the filter
plt.figure(figsize=(10, 6))
plt.plot(0.5 * sampling_frequency * w / np.pi, np.abs(h), 'b')
plt.title("Frequency Response of Third-order Derivative Filter")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz
import pandas as pd

def central_difference_filter(window_size):
    return np.array([-1, 8, 0, -8, 1]) / (12 * window_size)

def apply_filter(signal, filter_coeffs):
    filtered_signal = np.zeros_like(signal)
    for i in range(len(signal)):
        for j in range(len(filter_coeffs)):
            if i - 2 + j >= 0 and i - 2 + j < len(signal):
                filtered_signal[i] += filter_coeffs[j] * signal[i - 2 + j]
    return filtered_signal

# Read ECG signal from CSV file
ECG_1 = pd.read_csv('ECG1.csv')

# Extract time and ECG signal values for signal 1
time_1 = ECG_1.iloc[:, 0].values
ecg_signal_1 = ECG_1.iloc[:, 1].values

# Define window size for the central difference filter
window_size = 1  # We divide by window size for normalization

# Define 4-point central difference filter coefficients
filter_coeffs = central_difference_filter(window_size)

# Apply central difference filter to the ECG signal
filtered_signal = apply_filter(ecg_signal_1, filter_coeffs)

# Compute frequency response of the filter
w, h = freqz(filter_coeffs, worN=8000)

# Plot original and filtered ECG signals
plt.figure(figsize=(12, 6))
plt.plot(time_1, ecg_signal_1, label='Original ECG Signal')
plt.plot(time_1, filtered_signal, label='Filtered ECG Signal (Central Difference)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('ECG Signal Before and After Filtering')
plt.legend()
plt.grid()

# Plot frequency response of the filter
plt.figure(figsize=(12, 6))
plt.plot(0.5 * 500 * w / np.pi, np.abs(h))
plt.title("Frequency Response of 4-Point Central Difference Filter")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid()

plt.show()
```
